```python
import yfinance as yf
import pandas as pd
```

```python
tesla = yf.Ticker("TSLA")
```

```python
tesla_data = tesla.history(period="max")
tesla_data.head()
```

```python
tesla_revenue = tesla.history(period = "max")
tesla_revenue.tail()
```

```python
import matplotlib.pyplot as plt
tesla_history = tesla.history(period="max")

def make_graph(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data["Close"], label="Close Price", color="purple")
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.show()

make_graph(tesla_history, "Tesla Stock Price")
```
